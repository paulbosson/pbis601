➢ The brute force method for finding frequent itemsets works as follows.
Enumerate and generate all possible 1-itemsets and 2-itemsets. There are
30 items, so there are 435 possible 2-itemsets totally. Check to see whether
each possible 1-itemset/2-itemset is frequent. Then enumerate and generate

see whether each possible 3-itemset is :frequent. Keep on doing so until
you see none of the possible k-itemsets is :frequent for some k, at which
point the brute force method terminates without generating (k+ 1 )-
itemsets.

```python
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import fpgrowth
from mlxtend.frequent_patterns import apriori, association_rules

def calculate_frequent_itemsSets(transactions):
  # Convert the transaction data to a one-hot encoded DataFrame
  te = TransactionEncoder()
  te_ary = te.fit(transactions).transform(transactions)
  df = pd.DataFrame(te_ary, columns=te.columns_)

  # Generate frequent itemsets using FP-growth algorithm
  start_time = time.time()
  frequent_itemsets_fp = fpgrowth(df, min_support=minimum_support, use_colnames=True)

  # Display the frequent itemsets
  print("Frequent Itemsets (FP-growth):")
  print(frequent_itemsets_fp)
  end_time = time.time()
  runtime = end_time - start_time
  print(f"Runtime: {runtime:.6f} seconds")

  # Generate association rules with minimum confidence
  rules_fp = association_rules(frequent_itemsets_fp, metric="confidence", min_threshold=minimum_confidence)
  print("\nAssociation Rules (FP-growth):")
  print(rules_fp)
  end_time = time.time()
  runtime = end_time - start_time
  print(f"Runtime: {runtime:.6f} seconds")

  # Generate frequent itemsets using Apriori algorithm
  start_time = time.time()
  frequent_itemsets = apriori(df, min_support=minimum_support, use_colnames=True)

  # Display the frequent itemsets and association rules
  print("Frequent Itemsets (Apriori):")
  print(frequent_itemsets)
  end_time = time.time()
  runtime = end_time - start_time
  print(f"Runtime: {runtime:.6f} seconds")

  # Generate association rules with minimum confidence
  rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=minimum_confidence)

  print("\nAssociation Rules (Apriori):")
  print(rules)
  end_time = time.time()
  runtime = end_time - start_time
  print(f"Runtime: {runtime:.6f} seconds")

calculate_frequent_itemsSets(amazon_transactions)
calculate_frequent_itemsSets(nike_transactions)
calculate_frequent_itemsSets(hw_transactions)
calculate_frequent_itemsSets(bestbuy_transactions)
calculate_frequent_itemsSets(kmart_transactions)
```
